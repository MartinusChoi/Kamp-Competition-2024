# Kamp Competition PreProcessed Data EDA

## Library Import

```python
import pandas as pd
import matplotlib.pyplot as plt

from kamp.preprocess import load_data, NanProcessor, CatFeatureEncoder, remove_outliers_by_isoforest, T_Testor
from kamp.preprocess import NAN_GRID, ENCODE_GRID

DATA_PATH = './data/경진대회용 주조 공정최적화 데이터셋.csv'
```

## Data PreProcess

```python
# Load Raw Data
data_configs = load_data(DATA_PATH)
data = data_configs['data']
numeric_features = data_configs['numeric_features']

# Nan Data Process
data = NanProcessor(nan_grid=NAN_GRID).process(data)

# Encoding
data = CatFeatureEncoder(encode_grid=ENCODE_GRID).process(data)

# Remove Outlier
data = remove_outliers_by_isoforest(data=data, outlier_rate=0.08)

# T-Test
t_test = T_Testor(p_threshold=0.05)
t_test_configs = t_test.test(data)
data = t_test.get_useful_data(data)

data = data.reset_index(drop=True)
```
