# Kamp Competition PreProcessed Data EDA

## Library Import

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns
import matplotlib as mpl

from kamp.preprocess import load_preprocessed_unscaled_data

DATA_PATH = './data/경진대회용 주조 공정최적화 데이터셋.csv'
```

## Data PreProcess

```python
data = load_preprocessed_unscaled_data(DATA_PATH)
```

## 'count'별 Fail 비율

```python
data['count_group'] = data['count'].apply(lambda x: x if x <= 10 else '11+')

grouped = data[data['passorfail'] == 1].groupby('count_group')['passorfail'].sum() / data['passorfail'].sum()

fig, ax = plt.subplots(figsize=(10,10))

explode = [0.1 if label != '11+' else 0.2 for label in grouped.index]

ax.pie(
    grouped,
    labels=grouped.index,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'edgecolor' : 'black', 'linewidth' : 1},
    textprops = {'fontsize' : 12, 'color' : 'black'},
    pctdistance=0.85,
    explode=explode
)
centre_circle = plt.Circle((0, 0), 0.30, fc='white')
fig.gca().add_artist(centre_circle)

plt.title("'count'별 Fail 비율 (1-10 and 11+)")

plt.legend(grouped.index, loc='upper left', bbox_to_anchor=(1, 1))

ax.axis('equal')
plt.tight_layout()
plt.show()
```

## Correlation Matrix

```python
try : 
    data = data.drop(columns=['count_group'])
except : pass

corr_mat = data.corr()

fig, axe = plt.subplots(figsize=(20, 20))
sns.heatmap(corr_mat,
            annot=True,
            cmap='Spectral_r',
            fmt='.2f',
            vmin=-1,
            vmax=1,
            center=0,
            annot_kws={'size' : 10, 'size' : 15},
            linewidths=0.5)

plt.title('Feature 별 상관관계', fontsize=15)
plt.show()
```

## KDE Plot per Features

```python
data_melted = data.melt(id_vars=['passorfail'], value_vars=list(data.drop(columns=['passorfail']).columns), 
                        var_name='feature', value_name='value')

mpl.rcParams['font.family'] = 'DejaVu Sans'


for feature in list(data.drop(columns=['passorfail']).columns):
    fig, axe = plt.subplots(figsize=(10, 10))

    sns.kdeplot(data = data_melted[data_melted['feature'] == feature], 
                x='value', 
                label=feature, 
                fill=True, 
                alpha=0.5, 
                hue='passorfail')
    
    # plt.yscale('log')
    plt.title(f"KDE Plot of '{feature}' (Log Scale)")
    plt.xlabel('Value', fontsize=15)
    plt.ylabel('Density (Log Scale)', fontsize=15)

    plt.savefig(f'./data/kde/{feature}.png', dpi=300)

    plt.show()
```
